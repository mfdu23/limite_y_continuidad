```python
import sympy as sp

t=sp.symbols('t')
y = sp.Function('x')(t)
```

```python
ode = sp.Eq(y.diff(t) + y, t)

sol_general = sp.dsolve(ode, y)

print(sol_general)
```

```python
t0=0
y0=1
cond_inicial={y.subs(t,t0):y0}

sol_particular = sp.dsolve(ode, y, ics=cond_inicial)

print(f"Soluci贸n particular con y({t0})={y0}: {sol_particular}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

sol_particular_func = sp.lambdify(t, sol_particular.rhs, 'numpy')

t_vals = np.linspace(0, 10, 400)
y_vals = sol_particular_func(t_vals)

plt.figure(figsize=(10, 6))
plt.plot(t_vals, y_vals, label=str(sol_particular.rhs))
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Soluci贸n Particular de la Ecuaci贸n Diferencial')
plt.legend()
plt.grid(True)
plt.show()
```
