```python
import sympy as sp 
```

```python
x=sp.symbols('x')

f=1/(x-1)
```

```python
lim_left=sp.limit(f,x,1,dir='-')
lim_right=sp.limit(f,x,1,dir="+")
```

```python
print(f"Limite de f(x) cuando x tiende a 1 por izquierda: {lim_left}")
print(f"Limite de f(x) cuando x tiende a 1 por derecha: {lim_right}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_vals=np.linspace(0,2,400)
y_vals=[1/(val-1) if val != 1 else None for val in x_vals]

plt.figure(figsize=(10,6))
plt.plot(x_vals,y_vals,label=r'$f(x) = \frac{1}{x-1}$')
plt.axvline(x=1, color='red', linestyle='--', label='x=1')
plt.ylim(-10, 10)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Visualiuzación de la discontinuidad en x=1')
plt.legend()
plt.grid()
plt.show()
```
